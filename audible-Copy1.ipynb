```python
df.author=df.author.str.replace('Writtenby:','')
#removing writtenby from column author
```

```python
df.narrator=df.narrator.str.replace('Narratedby:','')
```

```python
listofstring=[]
for x in df['time']:
    listofstring.extend(x.split())

set(listofstring)
#checking for unwanted words from the column time
```

```python
df['time']=df['time'].str.replace('hrs','hr')
#standardising spelling of hour
```

```python
df['time']=df['time'].str.replace('mins','min').str.replace('minute','min')
#standardising spelling of minute
```

```python
df.head()
```

```python
df['hours']=df.time.str.extract(r'(\d+) hr')
#extracting the number of hours
```

```python
df['minutes']=df.time.str.extract(r'(\d+) min')
#extracting the number of minutes
```

```python
df.minutes.isnull().sum()
```

```python
df.minutes=df.minutes.fillna('0')
```

```python
df.hours=df.hours.fillna('0')
```

```python
df['timeinminutes']=df.hours.astype(int)*60 + df.minutes.astype(int)
#converting the time in integer value
```

```python
df.drop(['hours','minutes'],axis=1,inplace=True)
```

```python
df.head(10)
```

```python
df.releasedate=pd.to_datetime(df.releasedate,format='%d-%m-%y')
#converting string column date to datetime
```

```python
df['num_ratings']=df['stars'].str.extract(r'(\d+) ratings').astype(float)
#extracting the number of ratings
```

```python
df['avg_rating']= df['stars'].str.extract(r'(\d+|\d+\.\d+) out of').astype(float)
#extracting the average rating
```

```python
df.head()
```

```python
df['language']= df.language.str.capitalize()
```

```python
df.info()
```

```python
list(df.price.unique())
```

```python
df['price']=df.price.str.replace(',','')
#removing commas from the price column
```

```python
df['price']=df.price.str.replace('Free','0').astype(float)
#converting price to float
```

```python
df.head()
```

```python
df['author']=df.author.str.replace(r'(\w)([A-Z])', r'\1 \2', regex= True)
#adding space between first and last name
```

```python
df['narrator']=df.narrator.str.replace(r'(\w)([A-Z])', r'\1 \2', regex= True)
#adding space between first and last name
```
